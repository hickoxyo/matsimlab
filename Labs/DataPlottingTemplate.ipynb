```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import odr

poly_model = odr.polynomial(2)

#First, import data from table. Make sure the file you read is in the same directory as this notebook.
results_df = pd.read_csv('BaTiO3_EnergyVolumeData.csv')
xdata = results_df['a (Angstroms)']  # Setting the x values
ydata = results_df['Total Energy (eV)']  # setting the y values

#Next, we want to fit a polynomial to the data. There are many options, but we'll use polyfit for now
coeff = np.polyfit(xdata, ydata, deg=2) #find polynomial coefficients
print(coeff) #print the polynomial coefficients

#In order to plot the polynomial fit, we'll need to use the fit to generate a set of x and y data points
x_fit = np.linspace(0,10,100) 
y_fit = np.poly1d(coeff)

```

```python
#In this next section, we define the plot itself.

plt.figure(figsize = (10,5)) # set the size of the figure
plt.scatter(xdata, ydata) # scatter plot of the data
plt.plot(x_fit,y_fit(x_fit)) #plot of the fitted polynomial

#axis parameters
plt.xlabel('lattice parameter ($\AA$)',fontsize=14) 
plt.ylabel('total energy (eV)',fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlim(3.3,3.8)
plt.ylim(-310, -308)
#Finally, we save the plot as a high-quality image for use in a report, paper, presentation, etc.
plt.savefig('BaTiO3_EnergyVolPlot.png', dpi = 600, bbox_inches = 'tight')
```
