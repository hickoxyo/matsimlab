```python
import matplotlib.pyplot as plt
from matplotlib import rcParamsDefault
import numpy as np
%matplotlib inline

# load data
energy, dos, idos = np.loadtxt('BTO_dos.dat', unpack=True)
fermi=12.074

# make plot
plt.figure(figsize = (12, 6))
plt.plot(energy, dos, linewidth=0.75, color='red')
plt.yticks([])
plt.xlabel('Energy (eV)')
plt.ylabel('DOS')
plt.axvline(x=fermi, linewidth=0.5, color='k', linestyle=(0, (8, 10)))
plt.xlim(-6, 16)
plt.ylim(0, )
plt.fill_between(energy, 0, dos, where=(energy < fermi), facecolor='red', alpha=0.25)
plt.text(fermi-0.5, 1.7, 'Fermi energy', fontsize= 12, rotation=90)
plt.show()

#plt.savefig('BaTiO3_DOSPlot.png', dpi = 600, bbox_inches = 'tight')
```
